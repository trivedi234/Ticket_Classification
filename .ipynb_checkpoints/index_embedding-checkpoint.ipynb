```python
reduced_vectors = pd.read_csv('reduced_vectors.csv',header=None)
reduced_vectors=reduced_vectors.values
```

```python
data_set=np.zeros((2403279,3))
data_set[:,0]=dataS
data_set[721517:,1]=1
data_set[0:721517,2]=1
np.random.shuffle(data_set)
data_set
```

```python
for i in range(26394):
    reduced_vectors[i,:]=reduced_vectors[i,:]-np.min(reduced_vectors[i,:])
    reduced_vectors[i,:]=np.true_divide(reduced_vectors[i,:],np.max(reduced_vectors[i,:]))
```

```python
reduced_vectors
```

```python
data_index=0
def batchNext(size):
    global data_index
    index = data_set[data_index:data_index+size,0].astype(int)
    inputx = reduced_vectors[index-1,:]
    inputy = data_set[data_index:data_index+size,1:]
    data_index = (data_index+size)%2403279
    return inputx, inputy
trainx, trainy = batchNext(1682000)
valx, valy = batchNext(721278)
data_index=0
learning_rate = 0.3
epochs = 50
batch_size = 1000

graph = tf.Graph()

with graph.as_default():
    

    x = tf.placeholder(tf.float32, [None, 100])
    y = tf.placeholder(tf.float32, [None, 2])

    W1 = tf.Variable(tf.random_normal([100, 20], stddev=0.03), name='W1')
    b1 = tf.Variable(tf.random_normal([20]), name='b1')
    W2 = tf.Variable(tf.random_normal([20, 2], stddev=0.03), name='W2')
    b2 = tf.Variable(tf.random_normal([2]), name='b2')

    hidden_out = tf.add(tf.matmul(x, W1), b1)
    hidden_out = tf.nn.relu(hidden_out)
    y_ = tf.add(tf.matmul(hidden_out, W2), b2)
    
    
    y_clipped = tf.clip_by_value(y_, 1e-10, 0.9999999)
    cross_entropy=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=y_))
    optimiser = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cross_entropy)
    
    init = tf.global_variables_initializer()
    
    correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
    
with tf.Session(graph=graph) as session:
    init.run()
    print(session.run(accuracy, feed_dict={x: trainx, y: trainy}))
    total_batch = 1682
    for epoch in range(epochs):
            avg_cost = 0
            for i in range(total_batch):
                batch_x, batch_y = batchNext(batch_size)
                _, c = session.run([optimiser, cross_entropy], feed_dict={x: batch_x, y: batch_y})
                avg_cost += c / total_batch
            print("Epoch:", (epoch + 1), "cost =", "{:.3f}".format(avg_cost))
            data_index=0
    pred=tf.nn.softmax(y_)
    predy=pred.eval({x: testx})
    print(session.run(accuracy, feed_dict={x: trainx, y: trainy}))
    print(session.run(accuracy, feed_dict={x: valx, y: valy}))

    
    

```

```python
gender =list()
for i in range(len(predy)):
    if predy[i,0]>0.5:
        gender.extend('f')
    else:
        gender.extend('m')
```
